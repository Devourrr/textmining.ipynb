## KoNLPy를 사용한 한글 명사 빈도 계산

```python
from konlpy.tag import Okt
from collections import Counter

f = open('data.txt', 'r', encoding = 'utf-8')
text = f.read()
f.close()
# print(text)

# Okt 객체 생성
okt = Okt()
noun = okt.nouns(text)
# print(noun)

noun_list = [] # 한 글자 단어 제외한 단어들의 리스트
for v in noun:
    if len(v)> 1:
        noun_list.append(v)

count = Counter(noun_list)
# print(count)

# 명사 빈도 카운트
noun_list = count.most_common(100)
# for v in noun_list:
#     print(v)
    
# print(dict(noun_list))

# 워드 클라우드 생성
wc = wordcloud.WordCloud(font_path='C:\Windows\Fonts\malgun.ttf',
                          background_color='white',
                          width = 800, height = 600,
                         max_words = 100, max_font_size=450,
                         mask=mask,
                        colormap ='plasma')#.generate(keywords) # Blues, Reds, plasma...

cloud = wc. generate_from_frequencies(dict(noun_list))

# 워드 클라우드 출력
plt.imshow(cloud)
plt.axis('off')
plt.show()
wc.to_file('news.jpg')

```
